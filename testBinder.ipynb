Let's try to import libraties

```python
import mne
import matplotlib.pyplot as plt
mne.viz.set_browser_backend("matplotlib")
```

Now we read the data that is loaded on github

```python
RawData = (
    
    mne.io.read_raw_fif('student_data.fif')
    .load_data()
    .filter(l_freq=0.1, h_freq=40, picks='eeg')
    .notch_filter(50, picks='eeg')

)
```

now we can plot the psd

```python
RawData.compute_psd(fmax=80).plot(
         average=True, amplitude=False, picks="data", exclude="bads" );
```

finally we can ust plot the data as continuous signal

```python
%matplotlib widget
RawData.plot(
    duration=10,
    n_channels=60,
    scalings={"eeg": 100e-6},
    highpass=1, lowpass=30
);
```
