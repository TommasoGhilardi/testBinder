```python
import mne
import os
import matplotlib.pyplot as plt


sample_data_folder = mne.datasets.sample.data_path()
sample_data_raw_file = os.path.join(sample_data_folder, 'MEG', 'sample', 'sample_audvis_raw.fif')
raw = mne.io.read_raw_fif(sample_data_raw_file, preload=True)

# 2. Preprocessing (Basic Cleaning)
raw.filter(l_freq=0.1, h_freq=40)
raw.set_eeg_reference('average', projection=True)

# 3. Plotting the Power Spectral Density (PSD)
# We set fmax=60 to see the alpha/beta peaks clearly
print("--- Computing PSD ---")
spectrum = raw.compute_psd(fmax=60)
spectrum.plot(average=True, picks="data")

plt.show() # This opens the window on your computer
```
